```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip()
```

```python
df["Department"] = df["Department"].str.strip()
```

```python
df["Location"] = df["Location"].str.strip()
```

```python
df["Hire_Date"] = pd.to_datetime(df["Hire_Date"])
```

```python
df["Department"].value_counts().plot(kind="bar")
plt.title("Employees by Department")
plt.xlabel("Department")
plt.ylabel("Employees")
plt.show()
```

```python
df.groupby("Department")["Salary_INR"].mean().plot(kind="bar")
plt.title("Average Salary by Department")
plt.show()
```

```python
df["Work_Mode"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Work Mode Distribution")
plt.ylabel("")
plt.show()
```

```python
df["Performance_Rating"].value_counts().sort_index().plot(kind="bar")
plt.title("Performance Rating")
plt.show()
```

```python
df["Year"] = df["Hire_Date"].dt.year

df["Year"].value_counts().sort_index().plot(kind="line", marker="o")
plt.title("Hiring Trend")
plt.show()
```

```python
df["Location"].value_counts().plot(kind="bar")
plt.title("Employees by Location")
plt.show()
```
